```python
import os
from pydub import AudioSegment
import argparse
```

```python
def split_audio(input_folder, output_folder, segment_length):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for filename in os.listdir(input_folder):
        if filename.endswith(('.mp3', '.wav', '.flac', '.ogg')):
            file_path = os.path.join(input_folder, filename)
            audio = AudioSegment.from_file(file_path)
            duration = len(audio) 
            segment_length_ms = segment_length * 1000  
            
            num_segments = duration // segment_length_ms 
            
            base_name, ext = os.path.splitext(filename)
            
            for i in range(num_segments):
                start_time = i * segment_length_ms
                end_time = start_time + segment_length_ms
                segment = audio[start_time:end_time]
                output_filename = f"{base_name}_{i+1}{ext}"
                output_path = os.path.join(output_folder, output_filename)
                segment.export(output_path, format=ext[1:])
                print(f"Saved: {output_filename}")
```

```python
input_folder = r"C:\Users\Hp\Desktop\Lecture Files\DH604\Voice Samples\Cleaned - Italian-'a's\PD_silence_removed"
output_folder = r"C:\Users\Hp\Desktop\Lecture Files\DH604\mRMR\Italian a\PD"
split_audio(input_folder, output_folder, 2)
```
